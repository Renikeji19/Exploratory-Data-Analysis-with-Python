```python
import numpy as np
import pandas as pd
```

```python
Fao_data = pd.read_csv(r'C:\Users\HP840 G2\Downloads\FoodBalanceSheetsNew.csv')
Fao_data.head()
```

```python
#QUESTION 1: What is the total sum of Animal Fat produced in 2014 and 2017 respectively?
# ANSWER = 209460.54 and 269617.53


Fao_data.groupby(['Item'])[['Y2014','Y2017']].sum().loc['Animal fats']
```

```python
#QUESTION 2 : What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
#ANSWER : 135.236 and 1603.404

print(round(Fao_data['Y2015'].mean(), 3))
print(round(Fao_data['Y2015'].std(), 3))
```

```python
#QUESTION 3 : What is the total number and percentage of missing data in 2016 to 2 decimal places?
#ANSWER = 1535 and 2.52

#Total missing data in 2016
print("Total number: ",Fao_data['Y2016'].isnull().sum())

print("Percentage: %.2f" % round(Fao_data['Y2016'].isnull().sum().sum()/len(Fao_data['Y2016'])*100, 2))
```

```python
#QUESTION 4 : Which year had the highest correlation with ‘Element Code’? 
#ANSWER = 2014

corrData = Fao_data.corr(method='pearson')
corrData.loc[corrData['Element Code'].idxmax()]
```

```python
#QUESTION 5 : What year has the highest sum of Import Quantity?
#ANSWER = 2017

Fao_data.groupby(['Element']).sum().reset_index()
```

```python
#QUESTION 6 : What is the total number of the sum of Production in 2014?
#ANSWER = 1931287.75

Fao_data.groupby(['Element'])['Y2014'].sum().reset_index()
```

```python
#QUESTION 7 : Which of these elements had the highest sum in 2018?
#ANSWER = Domestic supply quantity

Fao_data.groupby(['Element'])['Y2018'].sum().reset_index()
```

```python
#QUESTION 8 : Which of these elements had the 3rd lowest sum in 2018?
#ANSWER = Protein supply quantity (g/capita/day)

Fao_data.groupby(['Element'])['Y2018'].sum().reset_index()
```

```python
#QUESTION 9 : What is the total Import Quantity in Algeria in 2018?
#ANSWER : 36238.29

allData = Fao_data.groupby(['Element','Area'])['Y2018'].sum().reset_index()
allData[allData['Area'] == 'Algeria']
```
